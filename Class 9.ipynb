```python
import pandas as pd
```

```python
df = pd.read_csv(path, 
                 header = None)
df
```

```python
## Data pre-processing
```

```python
transaction = []
for i in range(0, df.shape[0]): #for each row
    for j in range(0, df.shape[1]): # for each column
        transaction.append(df.values[i][j]) # adding each element to the empty set
transaction
```

```python
ndf = pd.DataFrame(transaction,
                   columns = ['items']
                  )
ndf.head(2)
```

```python
pip install numpy
```

```python
import numpy as np
```

```python
transaction = np.array(transaction)
transaction
```

```python
ndf = pd.DataFrame(transaction,
                   columns = ['items']
                  )
ndf.head(2)
```

```python
to_be_deleted = ndf[ndf['items']=='nan'].index
to_be_deleted
```

```python
ndf.drop(to_be_deleted,
         inplace = True)
ndf ['count'] = 1
ndf
```

```python
df_table = ndf.groupby('items').sum().sort_values('count',
                                                  ascending = False).reset_index()
df_table
```

```python
df_table ['all'] = 'x'
df_table
```

```python
tra = []
for i in range(0, df.shape[0]):
    tra.append([ str(df.values[i,j])
                for j in range (0, df.shape[1])
               ]
              )
tra
```

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(tra).transform(tra)
encoded_df = pd.DataFrame (te_ary,
                           columns = te.columns_)
encoded_df
```

```python
encoded_df.drop(['nan'], axis = 1, inplace = True)
encoded_df.columns
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
f_itemsets = fpgrowth(encoded_df,
                     min_support = 0.6,
                     use_colnames = True
                    )
f_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
a_rules = association_rules(f_itemsets,
                            metric = 'lift',
                            min_threshold = 1.5
                           )
a_rules
```
